```python
%matplotlib notebook
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import integrate
from LPModel import *
import pandas as pd

#df = pd.read_csv('Data', header=None)
#X = df.loc[0:10, [14]].values

def ReadData(path):
    path = 'ExampleAlexGitHub/Li-Petrasso/'+ path
    data = open(path, "r")
    lines = []
    for line in data:
        lines.append(line)   
    X = []
    Y = []      
    for j in range(0,len(lines)):
        string = lines[j].split(',')
        X.append(float(string[0]))
        Y.append(-float(string[1]))
    return X, Y
X, Y = ReadData('ProtonDT4.182gcc5keV.csv')
PEnergy1, StopPow1 = LP(1, 0,FieldDensity=4.182, FieldTem=5000, FieldCharge=1, FieldMass=2.5, ProjE=11, ProjM=1, ProjC=1 )
plt.loglog(PEnergy1, StopPow1, label='QuantumOn')
plt.loglog(X, Y, label='Alex')
plt.xlim(0.1, 11)
plt.ylim(0.001, 1)
plt.legend()
```
